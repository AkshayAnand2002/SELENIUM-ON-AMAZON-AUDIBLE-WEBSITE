```python
#HANDLING PAGINATION
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.options import Options
options = Options()
options.headless=False #to activate headless mode.
#options.add_argument('window-size=1920x1080')  # Set a big window size, so all the data will be displayed
#to assign size to screen showing output of headless mode as
#many times driver.maximize_window() does not work well for headless mode.
web = "https://www.audible.in/search"
path = "C:\chromedriver.exe"
driver = webdriver.Chrome(path , options=options)
driver.get(web)
driver.maximize_window()#window is spread all over the screen.
#pagination
pagination=driver.find_element_by_xpath('//ul[contains(@class,"pagingElements")]')# this represents the tag that contains 
#all page no. elements together as one that is seen or present at the bottom-right side of the webpage.
pages = pagination.find_elements_by_tag_name('li')#li tags inside pagination represents individual page nos 1, 2 ,3,...
#and not as all page nos together as one.the above is equivalent to-- //ul[contains(@class,"pagingElements")]/li
#pages[0] represents the first element of pages class i.e. left pointing arrow that helps in going to previous page,
#as seen from code and website.
#pages[1] represents 1st page , pages[2] represents 2nd page. pages[3] represents ... i.e. 3 dots as seen on website.
#pages[4] represents 60th page i.e. last page. pages[5] represents right pointing arrow which helps in moving to next page.
#for last element we can also use pages[-1], for previous -2 and so on.
#sincec last element is right arrow and last page is 60 is -2.
last_page = int(pages[-2].text) # int(pages[-2].text) represents or is 60.
#but it should not click beyond last page i.e. 60 .
current_page=1
book_title=[]
book_author=[]
book_length=[]
while current_page <= last_page:
    container = driver.find_element_by_class_name('adbl-impression-container')
    products = container.find_elements_by_xpath('./li')
    #. represents whole container i.e. container = driver.find_elements_by_class_name('adbl-impression-container')
    #/li represents immediate li element.
    #./li used as the class adbl-impression-container contains all its elements in li class which is seen by the inspection of website.
    #Using x-path theory we write //h3[contains(@class,"bc-heading")] to represent heading of search results.
    #h3 class contains headings of search results as seen by inspection of website.
    #below taking empty lists to store the data.
    for product in products: #.// used for specifying current context i.e. product here.
        book_title.append(product.find_element_by_xpath('.//h3[contains(@class,"bc-heading")]').text)
        #product.find_element_by_xpath('.//h3[contains(@class,"bc-heading")]').text
        #above code as heading is present in class contains letters or name "bc-heading".
        book_author.append(product.find_element_by_xpath('.//li[contains(@class,"authorLabel")]').text)
        #product.find_element_by_xpath('.//li[contains(@class,"authorLabel")]').text
        #above code as author name is present in li class named as or containing "authorLabel".
        book_length.append(product.find_element_by_xpath('.//li[contains(@class,"runtimeLabel")]').text)
        #product.find_element_by_xpath('.//li[contains(@class,"runtimeLabel")]').text
        #above code as length is present in li class named as or containing "runtimeLabel".
    current_page=current_page+1
    #next page i.e. right arrow on the website is in span class named or containing nextButton.
    try:
        next_page = driver.find_element_by_xpath('//span[contains(@class,"nextButton")]')
        next_page.click()#nextpage button has click functionality.
    except:
        pass
driver.quit()
#all required things placed inside while loop as we need to keep check of current page value and check while loop cond. also.
df_books = pd.DataFrame({'title':book_title,'author':book_author,'length':book_length})
df_books.to_csv('books_pagination.csv',index=False)
print(df_books)
```

```python
#PROBLEM OF JAVASCRIPT DRIVEN WEBSITES IS DATA IS LOADED DYNAMICALLY AND TAKES TIME FOR LOADING ELEMENTS.
#AN ELEMENT MIGHT NOT BE PRESENT WHEN SCRAPING A WEBSITE.
#SO DRIVER IS REQUIRED TO WAIT UNTIL ALL THE DATA WE WISH TO SCRAPE IS LOADED COMPLETELY.
#2 TYPES OF WAITS--
#1)IMPLICIT WAIT--TELL DRIVER TO WAIT FOR CERTAIN AMOUNT OF TIME WHEN TRYING TO LOCATE AN ELEMENT. EG- time.sleep(2)
#2)EXPLICIT WAIT-- TELL WEBDRIVER TO WAIT FOR A SPECIFIC CONDITION TO OCCUR BEFORE PROCEEDING FURTHER WITH EXECUTON.
#EG- WebDriverWait(driver,10).until(...)--we want driver to wait for 10 sec.inside i.e. wait until we some given condition
#happens and wait for 10 seconds.If after 10 seconds the condition is not satisfied the code is going to break.
#syntax of condition inside the until condition-- EC.presence_of_element_located(...) or EC.element_to_be_selected(...) etc.
#EC stands for expected condition. Inside the EC.something() brackets is (By.Xpath,"value")
```

```python
#WAITS
from selenium import webdriver
#new libraries for waits
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
web = "https://www.audible.in/adblbestsellers?ref=a_search_t1_navTop_pl1cg0c1r0&pf_rd_p=45682ad3-a4ca-473f-b9dc-aedd02f1e5fb&pf_rd_r=3DRQD4YYXDR3Q14FVMXT"
path = 'C:\chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get(web)
driver.maximize_window()

pagination = driver.find_element_by_xpath('//ul[contains(@class, "pagingElements")]')
pages = pagination.find_elements_by_tag_name('li')
last_page = int(pages[-2].text)

book_title = []
book_author = []
book_length = []
current_page = 1

while current_page <= last_page:
    # Implicit Wait
    #time.sleep(2)
    # Explicit Wait
    # container = driver.find_element_by_class_name('adbl-impression-container ') #WE REWRITE THIS LINE USING EXPLICIT WAITS BELOW.
    #only 1 element
    container = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, 'adbl-impression-container ')))
    # products = container.find_elements_by_xpath('./li')#WE REWRITE THIS LINE USING EXPLICIT WAITS BELOW.
    #many elements
    products = WebDriverWait(container, 5).until(EC.presence_of_all_elements_located((By.XPATH, './li')))

    for product in products:
        book_title.append(product.find_element_by_xpath('.//h3[contains(@class, "bc-heading")]').text)
        book_author.append(product.find_element_by_xpath('.//li[contains(@class, "authorLabel")]').text)
        book_length.append(product.find_element_by_xpath('.//li[contains(@class, "runtimeLabel")]').text)

    current_page = current_page + 1
    try:
        next_page = driver.find_element_by_xpath('.//span[contains(@class , "nextButton")]')
        next_page.click()
    except:
        pass

driver.quit()

df_books = pd.DataFrame({'title': book_title, 'author': book_author, 'length': book_length})
df_books.to_csv('books_pagination_explicit_wait.csv', index=False)
#here no print statement given.
```
